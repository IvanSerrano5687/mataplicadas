```python
#1. Serie Fibonacci
import matplotlib.pyplot as plt

def fibonacci(n):
    fib = [0, 1]
    for i in range(2, n):
        fib.append(fib[-1] + fib[-2])
    return fib

n = int(input("Introduce el número de términos de la serie Fibonacci: "))
fib_series = fibonacci(n)

print(f"Serie Fibonacci: {fib_series}")

plt.plot(fib_series, marker='o')
plt.title("Serie Fibonacci")
plt.xlabel("Índice")
plt.ylabel("Valor")
plt.grid()
plt.show()
```

```python
#2. Análisis de datos meteorológicos
import numpy as np
import matplotlib.pyplot as plt

temperaturas = [25.6, 26.1, 24.8, 27.3, 28.4, 26.7, 25.9, 29.1, 30.2, 27.8, 24.3, 23.9,
                26.5, 28.0, 27.4, 29.3, 30.5, 26.8, 25.7, 24.6]

media = np.mean(temperaturas)
mediana = np.median(temperaturas)
desviacion = np.std(temperaturas)

print(f"Media: {media:.2f}, Mediana: {mediana:.2f}, Desviación estándar: {desviacion:.2f}")

plt.plot(temperaturas, marker='o', label="Temperatura")
plt.axhline(media, color='r', linestyle='--', label=f"Media ({media:.2f})")
plt.title("Temperaturas Mensuales")
plt.xlabel("Mes")
plt.ylabel("Temperatura (°C)")
plt.legend()
plt.grid()
plt.show()
```

```python
#3. Simulación de un dado
import random
import matplotlib.pyplot as plt

lanzamientos = 1000
resultados = [random.randint(1, 6) for _ in range(lanzamientos)]

frecuencia = [resultados.count(i) for i in range(1, 7)]

print(f"Frecuencia de cada número: {frecuencia}")

plt.bar(range(1, 7), frecuencia, color='blue', alpha=0.7)
plt.title("Frecuencia de lanzamientos de un dado")
plt.xlabel("Número del dado")
plt.ylabel("Frecuencia")
plt.grid(axis='y')
plt.show()
```

```python
#4. Gráfica de funciones trigonométricas
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 100)
sin_x = np.sin(x)
cos_x = np.cos(x)
tan_x = np.tan(x)

plt.plot(x, sin_x, label="sin(x)")
plt.plot(x, cos_x, label="cos(x)")
plt.plot(x, tan_x, label="tan(x)", linestyle='--', alpha=0.7)
plt.ylim(-2, 2)  # Limitar el eje Y para tan(x)
plt.title("Funciones Trigonométricas")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
```

```python
#5. Raíces cuadradas
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
sqrt_x = np.sqrt(x)

plt.plot(x, sqrt_x, label="y = √x")
plt.plot(x, x, label="y = x", linestyle='--')
plt.title("Raíz Cuadrada y Línea Recta")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

```python
#6. Pendiente de una recta
import matplotlib.pyplot as plt

# Solicitar las coordenadas de los puntos al usuario
x1, y1 = map(float, input("Introduce las coordenadas del primer punto (x1, y1) separadas por un espacio: ").split())
x2, y2 = map(float, input("Introduce las coordenadas del segundo punto (x2, y2) separadas por un espacio: ").split())

# Calcular la pendiente de la recta
if x2 - x1 != 0:  # Evitar división por cero
    pendiente = (y2 - y1) / (x2 - x1)
    print(f"La pendiente de la recta es: {pendiente:.2f}")
else:
    print("La pendiente es indefinida (recta vertical).")

# Graficar la recta
plt.plot([x1, x2], [y1, y2], marker='o', label=f"Recta (Pendiente = {pendiente:.2f})" if x2 - x1 != 0 else "Recta vertical")
plt.title("Gráfica de la Recta")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')  # Eje X
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')  # Eje Y
plt.grid()
plt.legend()
plt.show()
```

```python
#7. Movimiento parabólico
import numpy as np
import matplotlib.pyplot as plt

v = float(input("Introduce la velocidad inicial (m/s): "))
angulo = float(input("Introduce el ángulo de lanzamiento (grados): "))

g = 9.81
angulo_rad = np.radians(angulo)

t_vuelo = 2 * v * np.sin(angulo_rad) / g
t = np.linspace(0, t_vuelo, num=500)

x = v * np.cos(angulo_rad) * t
y = v * np.sin(angulo_rad) * t - 0.5 * g * t**2

plt.plot(x, y)
plt.title("Movimiento Parabólico")
plt.xlabel("Distancia (m)")
plt.ylabel("Altura (m)")
plt.grid()
plt.show()
```

```python
# 8. Potencia de una función con deslizador
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
import warnings

warnings.filterwarnings("ignore", category=UserWarning)

def graficar_potencia(n):
    x = np.linspace(1, 10, 100)
    y = x**n
    plt.plot(x, y, label=f"x^{n}")
    plt.title(f"Gráfica de x^{n}")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid()
    plt.legend()
    plt.show()

interact(graficar_potencia, n=(1, 10));
```
